```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
import plotly.express as px
import plotly.graph_objects as go
```

```python
grouped=df.groupby('title').agg({'average_rating':[np.size,np.sum,np.mean]})

#agg toplama yapıyor.
#gruplara ayırdık
```

```python
grouped

#size= not veren kişi sayısı - 1 kişi 
#sum= rating toplamaı - 4,22
#mean= rating ortalaması - 4,22
# en az toplam 100 kişi yorum yapmış olamalı ve rating ortalaması yüksek olmalı ki tavsiye edelim
```

```python
popular=grouped.sort_values(('average_rating','mean'),ascending=False)

# grupladığımız lsitedeki ortalama ratinge göre sıralama
```

```python
popular.head()
```

```python
toplam=grouped['average_rating']['sum'].sum()
```

```python
toplam
#verilen bütün ratinglerin toplamı
```

```python
popular['percentage']=popular['average_rating','sum'].div(toplam)*100
```

```python
popular.head()

#ortalama (percentage) adında  yeni bir sütun oluşturduk
```

```python
popular.sort_values('percentage',ascending=False)
# ortalamaya göre büyükten küçüğe göre sıraladık
```
